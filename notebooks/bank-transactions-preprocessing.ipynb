```python
#Data Cleaning
# Removing unnecessary columns like '.' and'CHEQ NO.' (has missing values)


df.drop(columns = ['.','CHQ.NO.'], inplace = True)
```

```python
# filling the missing values
df['TRANSACTION DETAILS'] = df['TRANSACTION DETAILS'].fillna('Unknown')
df['WITHDRAWAL AMT'] = df['WITHDRAWAL AMT'].fillna(0)
df['DEPOSIT AMT'] = df['DEPOSIT AMT'].fillna(0)
```

```python
# Rename the columns
df.columns = [
    'account_no',
    'transaction_date',
    'transaction_details',
    'value_date',
    'withdrawal_amt',
    'deposit_amt',
    'balance_amt'
]
```

```python
#Create year, month, day
df['year'] = df['transaction_date'].dt.year
df['month'] = df['transaction_date'].dt.month
df['day'] = df['transaction_date'].dt.day
```

```python
#create transaction type
df['transaction_type'] = df.apply(
    lambda x: 'Withdrawal' if x['withdrawal_amt'] > 0 else 'Deposit',axis = 1
)
```

```python
df['net_transaction_amt'] = (df['deposit_amt']- df['withdrawal_amt'])
```

```python
def categorize_transaction(detail):

    detail = str(detail).upper()

    # NEFT
    if 'NEFT' in detail or 'NATIONAL ELECTRONIC' in detail:
        return 'NEFT'

    # RTGS
    elif 'RTGS' in detail or 'REAL TIME GROSS' in detail:
        return 'RTGS'

    # IMPS
    elif 'IMPS' in detail:
        return 'IMPS'

    # Cash
    elif 'CASH' in detail:
        return 'Cash'

    # Fund Transfer
    elif 'FUND' in detail:
        return 'Fund Transfer'

    # Transfer
    elif 'TRF' in detail or 'TRANSFER' in detail:
        return 'Transfer'

    # ATM/NFS
    elif 'NFS' in detail or 'ATM' in detail:
        return 'ATM/NFS'

    # Charges & Tax
    elif 'TAX' in detail or 'CHARGE' in detail:
        return 'Charges'

    # International
    elif 'USD' in detail or 'TRANGLO' in detail:
        return 'International Transfer'

    # Internal Banking
    elif 'FDRL' in detail:
        return 'Internal Banking'

    else:
        return 'Others'
```

```python
df['transaction_category']=(df['transaction_details'].apply(categorize_transaction))
```
